```python
#연령대별
```

```python
# '나이' 열을 기반으로 연령대를 구분합니다. 예를 들어, 20대(20-29), 30대(30-39), 40대(40-49) 등으로 구분합니다.
df_kor['연령대'] = pd.cut(df_kor['나이'], bins=[19, 29, 39, 49, 59, 69, 79, 89, 99], labels=['20대', '30대', '40대', '50대', '60대', '70대', '80대', '90대'])
```

```python
# 위암 환자와 위암 생존자 데이터 필터링
df_kor_patients = df_kor[df_kor['위암_환자'] == True]
df_kor_survivors = df_kor[df_kor['위암_생존자'] == True]

# 연령대별 위암 환자 및 생존자 수 계산
age_distribution_patients = df_kor_patients['연령대'].value_counts().sort_index()
age_distribution_survivors = df_kor_survivors['연령대'].value_counts().sort_index()

# 연령대별 분포를 위한 DataFrame 생성
age_df = pd.DataFrame({
    '연령대': age_distribution_patients.index,
    '환자 수': age_distribution_patients.values,
    '그룹': '위암 환자'
})
age_df_survivors = pd.DataFrame({
    '연령대': age_distribution_survivors.index,
    '환자 수': age_distribution_survivors.values,
    '그룹': '위암 생존자'
})
age_df_combined = pd.concat([age_df, age_df_survivors])

# 연령대별 분포 시각화 (라인 플롯)
plt.figure(figsize=(12, 6))
sns.lineplot(data=age_df_combined, x='연령대', y='환자 수', hue='그룹', style='그룹', markers=True, dashes=False)
plt.title('연령대별 위암 환자와 생존자 분포')
plt.xlabel('연령대')
plt.ylabel('인원 수')
plt.legend(title='그룹')
plt.show()
```

```python
# 남성 위암 환자 및 생존자 데이터 필터링
male_patients = df_kor[(df_kor['성별'] == 1) & (df_kor['위암_환자'] == True)]
male_survivors = df_kor[(df_kor['성별'] == 1) & (df_kor['위암_생존자'] == True)]

# 여성 위암 환자 및 생존자 데이터 필터링
female_patients = df_kor[(df_kor['성별'] == 2) & (df_kor['위암_환자'] == True)]
female_survivors = df_kor[(df_kor['성별'] == 2) & (df_kor['위암_생존자'] == True)]

# 연령대별 환자 및 생존자 수 계산
male_age_distribution_patients = male_patients['연령대'].value_counts().sort_index()
male_age_distribution_survivors = male_survivors['연령대'].value_counts().sort_index()
female_age_distribution_patients = female_patients['연령대'].value_counts().sort_index()
female_age_distribution_survivors = female_survivors['연령대'].value_counts().sort_index()

# 데이터 시각화
male_age_df = pd.DataFrame({
    '연령대': male_age_distribution_patients.index,
    '남자_위암 환자': male_age_distribution_patients.values,
    '남자_위암 생존자': male_age_distribution_survivors.values,
})

female_age_df = pd.DataFrame({
    '연령대': female_age_distribution_patients.index,
    '여자_위암 환자': female_age_distribution_patients.values,
    '여자_위암 생존자': female_age_distribution_survivors.values,
})

# 연령대별 분포 시각화 (라인 플롯)
plt.figure(figsize=(12, 6))
sns.lineplot(data=male_age_df, x='연령대', y='남자_위암 환자', label='남자_위암 환자', marker='o', color='blue')
sns.lineplot(data=male_age_df, x='연령대', y='남자_위암 생존자', label='남자_위암 생존자', marker='o', color='blue')
sns.lineplot(data=female_age_df, x='연령대', y='여자_위암 환자', label='여자_위암 환자', marker='o', color='red')
sns.lineplot(data=female_age_df, x='연령대', y='여자_위암 생존자', label='여자_위암 생존자', marker='o', color='red')
plt.title('성별에 따른 연령대별 위암 환자와 생존자 수')
plt.xlabel('연령대')
plt.ylabel('인원 수')
plt.legend(title='그룹')
plt.show()
```
